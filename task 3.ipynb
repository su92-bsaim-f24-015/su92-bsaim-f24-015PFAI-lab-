```python


MAX_X = int(input("Enter capacity of Jug X: "))
MAX_Y = int(input("Enter capacity of Jug Y: "))
GOAL = int(input("Enter the goal amount: "))

x = 0
y = 0

def show_state():
    print(f"\nCurrent State: X = {x} | Y = {y}")

def check_goal():
    if x == GOAL or y == GOAL:
        print("\n🎉 Congratulations! You reached the goal!")
        return True
    return False

while True:
    show_state()
    
    print("\nChoose an action:")
    print("1. Fill X")
    print("2. Fill Y")
    print("3. Empty X")
    print("4. Empty Y")
    print("5. Pour X into Y")
    print("6. Pour Y into X")
    print("7. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        x = MAX_X

    elif choice == "2":
        y = MAX_Y

    elif choice == "3":
        x = 0

    elif choice == "4":
        y = 0

    elif choice == "5":
        transfer = min(x, MAX_Y - y)
        x -= transfer
        y += transfer

    elif choice == "6":
        transfer = min(y, MAX_X - x)
        y -= transfer
        x += transfer


    elif choice == "7":
        print("Game exited.")
        break

    else:
        print("Invalid choice! Try again.")
        continue

    if check_goal():
        show_state()
        break
```
